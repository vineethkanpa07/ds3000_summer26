## Standard of Living in the EU
Here, standard of living will be measued using the Human Development Index (HDI). It is a grading scheme developed by the United Nations and it focuses on education, life expectancy, and people rather than just economic output

![Image of the EU](https://maproom.net/wp-content/uploads/EU-simple-map-preview.png)


Heres a [list of all European Union members by HDI](https://www.theglobaleconomy.com/rankings/human_development/European-union/)

### Factors HDI uses to calculate score
1. Life Expectancy Index
2. Education Index
3. Income Index

Final Score is geometric mean of these three indecies. For a detailed explenation of how these indecies work please see [the calculation part of this page](https://en.wikipedia.org/wiki/Human_Development_Index#New_method_(2010_HDI_onwards))
