# EU Climate Change Goals
Current European climate law aims to achieve the goals outlined in the European Green New Deal, for Europe’s economy and society to become climate-neutral by 2050.

![EU Green New Deal Goals Breakdown](https://www.researchgate.net/profile/Maria-Nogal/publication/343180791/figure/fig2/AS:916749243412483@1595581460335/Policy-areas-and-key-actions-relevant-to-the-construction-sector-as-part-of-the-European.png)

## Implementation Efforts
The European Commission assessed progress towards climate neutrality and adaptation objectives for the first time in 2023. While greenhouse gas levels are reducing, they reported that so far progress has been ineffiecient and more actions need to be taken to reduce emissions.

2023 Recommendations for the consistency of national measures with the climate-neutrality objective per Country
| Belgium | Bulgaria | Ireland |
| --------| --------- | -------- |
| Clarify Belgium’s long-term climate-neutrality goal | Submit the draft update of its latest notified integrated national energy and climate plan | Step up climate mitigation efforts |
| Substantiating Belgium’s emission reductions | Update and increase the ambition and quality on the national long-term strategy | Reduce emissions in agriculture |

### Key Features of the Climate Law
- a commitment to negative emissions after 2050
- a commitment to engage with sectors to prepare sector-specific roadmaps charting the path to climate neutrality in different areas of the economy
- stronger provisions on adaptation to climate change
- strong coherence across Union policies with the climate neutrality objective
- the establishment of the [European Scientific Advisory Board on Climate Change](https://climate-advisory-board.europa.eu/) to provide independent scientific advice
