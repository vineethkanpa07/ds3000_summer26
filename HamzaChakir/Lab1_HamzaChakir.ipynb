## Short History of the European Union

### Modern Role in International Politics

The European Union (EU) was created after World War II to help European countries cooperate and avoid future wars. What began as a small economic partnership between countries later grew into a major political and economic union with 27 member states.

![European Union Flag](https://upload.wikimedia.org/wikipedia/commons/b/b7/Flag_of_Europe.svg)

Today, the EU helps countries work together on trade, travel, technology, climate policy, and international relations.

| Germany | Berlin | Euro |

Main goals of the European Union:

- Maintain peace in Europe
- Support economic cooperation
- Protect human rights
- Allow free movement between countries
