# The European Parliment

## Summary 

The European Parliment (EP) is directly elected legislative body of the European Union that represents citizens to ensure democratic order. Together, the Parliment works with the Council of the EU based on proposals from the European Commision and passes EU laws that span across different dimensions of the internal market, environment, and agriculture. Members of the European Parliament (MEPs). MEPs maintain ties with non-EU countries, promote democracy, monitor human rights, and approve international trade agreements.
