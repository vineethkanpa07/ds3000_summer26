### Tariffs
![Tariffs Image](https://liveindex.org/wp-content/uploads/2025/03/US-Tariffs.jpg)
## What are tariffs?
Tariffs are taxes imposed by governments on imported goods, aimed at protecting domestic industries and influencing trade relationships.
## Types of tariffs
| Name | Definition |
|------------|------------------|
|Ad Valorem Tariff|A percentage-based tax on the value of imported goods|
|Specific Tariff|A fixed amount charged per unit of the imported product|
|Compound Tariff|A combination of both ad valorem and specific tariffs|
|Tariff-rate Quota|A system that allows a set quantity of a product to be imported at a lower tariff rate, with higher tariffs applied to quantities exceeding that limit|
