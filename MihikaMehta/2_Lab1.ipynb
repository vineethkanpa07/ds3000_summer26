## Part 2.2: Exploring Arrays (15 points)

1. Give the shape, size, ndim, and nbytes for each of the two arrays.
1. Take the transpose of both arrays. Call these `t_array_a` and `t_array_b`.
1. Try to add `array_a` and `t_array_b` (*prove* and *show* you did this with commented out code), then remove the last column of `t_array_b` and try to add them again. In a markdown cell, explain what happened.

```python
#1. Shape, size, ndim, nbytes 
print("array_a:") 
print("shape: ", array_a.shape) 
print(" shape: ", array_a.size)
print(" ndim; ", array_a.ndim) 
print(" nbytes:", array_a.nbytes)

print("\narray_b:")
print("  shape: ", array_b.shape)
print("  size:  ", array_b.size)
print("  ndim:  ", array_b.ndim)
print("  nbytes:", array_b.nbytes)

#2. Transpose
t_array_a = array_a.T 
t_array_b = array_b.T  

print("\nt_array_a shape:", t_array_a.shape)
print("t_array_b shape:", t_array_b.shape)

## 3a. Tried to add array_a and t_array_b, it raised a ValueError because array_a is (3,4) and t_array_b is (3,5) the shapes dont match, print(array_a + t_array_b)

## 3b. Removing last column of t_array_b and then add. 

t_array_b_trimmed = t_array_b[:, :-1] #shape will become (3,4)
result = array_a + t_array_b_trimmed
print("\narray_a + t_array_b_trimmed:\n", result)

## What happened? 
#After removing the last column of t_array_b with [:, :-1], its shape becomes (3, 4) - mathcing array_a - so the addtion works and produces the ourput.
```

```python
# make sure to import pandas library
import pandas as pd

df = pd.read_csv('train_stations_europe.csv', index_col='id')
print(df.head())
```

## Part 3.2: Manipulating Data (20 points)

    - What Series did the `.describe()` function run on? What Series did it not run on? What is the difference, and what does this mean the `.describe()` function is used for?

```python

belgium_df = df[(df['country'] == 'BE') & (df['is_city'] == True)]
print(belgium_df.head())

belgium_df.describe()

## The describe() function ran on the numeric columns: 'latitude', 'longitude', and 'parent_station_id', It didnt run on th boolean columns ('is_city', 'is_main_station', 'is_airport') or the string/object columns ('name', 'country', 'time_zone'). 
#The difference is that numeric columns contain quantitative data that statistics like mean, standard deviation, min, and max are meaningful for. AString columns represent categories or labels, there is no average of country names. Boolean columns are also skipped since they are non-numeric. 
#This means '.describe()' is primarily used for expolratory data analysis on numerical data and gives a quick summary of the range, center, and spread of your data before doinf deeper analysis.
```
