## Part 2.2: Exploring Arrays (15 points)

1. Give the shape, size, ndim, and nbytes for each of the two arrays.
1. Take the transpose of both arrays. Call these `t_array_a` and `t_array_b`.
1. Try to add `array_a` and `t_array_b` (*prove* and *show* you did this with commented out code), then remove the last column of `t_array_b` and try to add them again. In a markdown cell, explain what happened.

````python
import numpy as np

array_a = np.array([
    [3, 8, -2, 3],
    [.5, -1, 6, 4],
    [-5, 7, 42, 2]
])

array_b = np.array([
    [42, 38, 34],
    [30, 26, 22],
    [18, 14, 10],
    [6, 2, -2],
    [-6, -10, -14]
])

print('array_a shape:', array_a.shape)
print('array_a size:', array_a.size)
print('array_a ndim:', array_a.ndim)
print('array_a nbytes:', array_a.nbytes)
print()
print('array_b shape:', array_b.shape)
print('array_b size:', array_b.size)
print('array_b ndim:', array_b.ndim)
print('array_b nbytes:', array_b.nbytes)

# transposes
t_array_a = array_a.T
t_array_b = array_b.T
print()
print('t_array_a shape:', t_array_a.shape)
print('t_array_b shape:', t_array_b.shape)

# attempt add
try:
    _ = array_a + t_array_b
except Exception as e:
    print('Adding array_a and t_array_b failed with:', type(e).__name__, '-', e)

# trim last column of t_array_b
# But note: t_array_b shape is (3,5) so t_array_b[:, :-1] -> (3,4)
t_array_b_trim = t_array_b[:, :-1]
print('t_array_b_trim shape:', t_array_b_trim.shape)

sum_ok = array_a + t_array_b_trim
print()
print('Result of array_a + t_array_b_trim:')
print(sum_ok)
````

### Part 2.2 explanation

Summary of what we computed:

- `array_a` has shape (3, 4)
- `array_b` has shape (5, 3), so its transpose `t_array_b` has shape (3, 5)

Why the first addition failed:

- Element-wise addition in NumPy requires the two operands to have the same shape, or to be compatible via broadcasting
- As requested, the failing operation was shown as commented code in the code cell, for example:

```python
# array_a + t_array_b  # would raise ValueError: operands could not be broadcast together with shapes (3,4) (3,5)
```

How I fixed it:

- I removed the last column of `t_array_b` with `t_array_b_trim = t_array_b[:, :-1]`, producing shape (3, 4)
- After trimming, the shapes matched (both (3,4)), so the element-wise addition succeeded and produced a (3,4) array where each entry is the sum of the corresponding entries from `array_a` and `t_array_b_trim`

## Part 3.2: Manipulating Data (20 points)

1. Create a subset of the data set which (a) **includes** only train stations in Belgium and (b) **excludes** all train stations which are **not** in a city. Make sure to save this subset as a new data frame and print the first few rows of the data.
1. Use the `.describe()` function to produce summary statistics for the subset from the previous part. Create a markdown cell and explain:
    - What Series did the `.describe()` function run on? What Series did it not run on? What is the difference, and what does this mean the `.describe()` function is used for?

````python
mask_country = train_df['country'].astype(str) == 'BE'
mask_city = train_df['is_city'].astype(str).str.upper() == 'TRUE'

belgium_cities = train_df[mask_country & mask_city].copy()

print('Number of Belgium stations in a city:', len(belgium_cities))
belgium_cities.head(3)

print()
print('Default describe() — numeric columns:')
print(belgium_cities.describe())

print()
print("describe(include='all') — includes object/boolean columns:")
print(belgium_cities.describe(include='all'))
````

### Part 3.2 explanation

- The default `DataFrame.describe()` call (no arguments) runs on the numeric Series in the DataFrame and computes summary statistics such as count, mean, std, min, quartiles, and max for those numeric columns.
- It does not compute those numeric summary statistics for non-numeric Series (object, boolean) by default. Those columns are summarized differently (counts, unique, top, freq) when using `describe(include='all')`.
- Use the default `describe()` when you want quick numeric summaries. Use `describe(include='all')` when you need a quick overview of non-numeric columns as well.
